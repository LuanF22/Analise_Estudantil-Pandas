# Análise Estudantil

```python
# Lendo as bases de dados e importando bibliotecas
import pandas as pd
tabela_materia = pd.read_excel('Tabela_1.xlsx')
tabela_professora = pd.read_excel('Tabela_2.xlsx')
tabela_estudante = pd.read_excel('Tabela_3.xlsx')
tabela_nota = pd.read_excel('Tabela_4.xlsx')
```

## Agrupamento da média de notas por matéria e semestre

```python
#Juntando a tabela de notas com a tabela das matérias
tabela_nota_join_materia = tabela_nota.merge(tabela_materia, left_on = 'id_materia', right_on = 'id') 

#Agrupando a média de notas por matéria por trimestre
tabela_nota_join_materia = tabela_nota_join_materia.groupby(['descricao', 'data_inicio_trimestre']).nota.agg(['mean'])

#Renomear a coluna mean para media e imprimindo
tabela_nota_join_materia.rename(columns={'mean': 'media'}) 

```

## Agrupamento da média de notas por aluno e trimestre

```python
#Juntando a tabela de notas com a tabela dos estudantes
tabela_nota_join_estudante = tabela_nota.merge(tabela_estudante, left_on = 'id_estudante', right_on = 'id') 

#Agrupando a média de notas por aluno por trimestre e reduzindo o número de casas decimais para 2
tabela_nota_join_estudante = tabela_nota_join_estudante.groupby(['nome', 'data_inicio_trimestre']).nota.agg(['mean']).round(decimals = 2)

#Renomear a coluna mean para media e imprimindo
tabela_nota_join_estudante.rename(columns={'mean': 'media'}) 

```

## Número de reprovações por professor

```python
#Juntando a tabela de notas com a tabela das materias (para acessar as notas minimas de cada uma)
tabela_nota_join_materia = tabela_nota.merge(tabela_materia, left_on = 'id_materia', right_on = 'id') 

#Agrupar e contar os dados, por id_professor, do número de reprovações (nota do aluno < nota mínima)
reprov = tabela_nota_join_materia.groupby(['id_professor(a)']).apply(lambda x: x[x['nota'] < x['nota_minima']]['nota'].count()).to_frame()

#Juntando a tabela que estava utilizando (agrupamento das reprovações por id_professor) com a tabela dos professores
reprov =  reprov.merge(tabela_professora, left_on = 'id_professor(a)', right_on = 'id') #juntar

#Retirando as colunas que não tenho interesse
reprov = reprov.drop(columns=['id', 'data_admissao', 'formacao']) 

#Renomeando a coluna 0 para reprovacoes e colocando em ordem crescente
reprov = reprov.rename(columns={0: 'reprovacoes'}).sort_values('reprovacoes') 

#Imprimir resultados
reprov
```

## Número de aprovações por professor

```python
#Juntando a tabela de notas com a tabela das materias 
tabela_nota_join_materia = tabela_nota.merge(tabela_materia, left_on = 'id_materia', right_on = 'id') 

#Agrupar e contar os dados, por id_professor, das aprovações (nota do aluno >= nota minima)
aprov = tabela_nota_join_materia.groupby(['id_professor(a)']).apply(lambda x: x[x['nota'] >= x['nota_minima']]['nota'].count()).to_frame()

#Juntando a tabela que estava utilizando (agrupamento das aprovações por id_professor) com a tabela dos professores
aprov =  aprov.merge(tabela_professora, left_on = 'id_professor(a)', right_on = 'id')

#Retirando as colunas que não tenho interesse
aprov = aprov.drop(columns=['id', 'data_admissao', 'formacao'])

#Renomeando a coluna 0 para aprovacoes e colocando em ordem crescente
aprov = aprov.rename(columns={0: 'aprovacoes'}).sort_values('aprovacoes') 

#Imprimir resultados
aprov
```

## Análise do desvio padrão de notas por curso

```python
#Juntando a tabela de notas com a tabela dos estudantes
tabela_nota_join_estudante = tabela_nota.merge(tabela_estudante, left_on = 'id_estudante', right_on = 'id') #juntar

#Agrupar o desvio padrão das notas por curso e renomeando a coluna std para Desvio Padrao 
tabela_nota_join_estudante = tabela_nota_join_estudante.groupby('curso').nota.agg(['std']).rename(columns={'std': 'Desvio Padrao'})

#Arredondando para 3 casas decimais e imprimindo
tabela_nota_join_estudante.round(decimals = 3)
```

## Análise do desvio padrão de notas por aluno

```python
#Juntando a tabela de notas com a tabela dos estudantes
tabela_nota_join_estudante = tabela_nota.merge(tabela_estudante, left_on = 'id_estudante', right_on = 'id') 

#Agrupar o desvio padrão das notas por aluno e renomeando a coluna std para Desvio Padrao
tabela_nota_join_estudante  = tabela_nota_join_estudante .groupby('nome').nota.agg(['std']).rename(columns={'std': 'Desvio Padrao'}) 

#Colocar as informações em ordem crescente, arredondando para 3 casas decimais e imprimindo
tabela_nota_join_estudante.sort_values('Desvio Padrao').round(decimals = 3)

```

## Análise do número de faltas por matéria

```python
#Juntando a tabela de notas com a tabela das matérias
tabela_nota_join_materia = tabela_nota.merge(tabela_materia, left_on = 'id_materia', right_on = 'id') 

#Agrupando a soma das faltas por matéria
tabela_nota_join_materia = tabela_nota_join_materia.groupby('descricao').faltas.agg(['sum'])

#Renomeando a columa sum para total_de_faltas e a coluna descricao para materia
tabela_nota_join_materia = tabela_nota_join_materia.rename(columns={'sum': 'total_de_faltas', 'descricao':'materia'})

#Colocando em ordem crescente e imprimindo
tabela_nota_join_materia.sort_values('total_de_faltas')
```

## Análise de qual aluno tem 25% de faltas

### Com os dados apresentados nas 4 tabelas, não é possível calcular a porcentagem de faltas por aluno. Para tal, precisariamos do número total de aulas de cada matéria ou quantidade de presença de cada aluno. Para conseguir prosseguir, vou considerar que cada matéria tem 75 aulas.

```python
#Criando uma variavel para alocar essa nova tabela
tabela_materia_aula = tabela_materia

#Adicionar uma coluna com a quantidade de aulas para cada matéria
tabela_materia_aula['aulas'] = 75 

#Juntar a tabela de notas com a tabela das matérias (com as aulas)
tabela_nota_join_materia_aula = tabela_nota.merge(tabela_materia_aula, left_on = 'id_materia', right_on = 'id') 

#Somar o total de faltas por estudante por matéria 
tabela_nota_join_materia_aula = tabela_nota_join_materia_aula.groupby(['id_estudante', 'descricao','aulas']).faltas.agg(['sum']).reset_index()

#Calculo da taxa de ausencia por estudante por matéria (faltas/total de aulas)
tabela_nota_join_materia_aula['taxa_de_ausencia(%)'] = (tabela_nota_join_materia_aula['sum']/tabela_nota_join_materia_aula['aulas'])*100

#Juntar a tabela dos estudantes à tabela que estamos usando (matérias + notas)
nome_estd = tabela_nota_join_materia_aula.merge(tabela_estudante, left_on = 'id_estudante', right_on = 'id') 

#Filtrar colunas para aparecer os dados que queremos
nome_estd = nome_estd.filter(items=['nome', 'descricao', 'taxa_de_ausencia(%)']) 

#Colocando em ordem decrescente da taxa de ausencia por aluno por matéria
nome_estd = nome_estd.sort_values('taxa_de_ausencia(%)', ascending=[False]).round(decimals = 2)

#Renomando a coluna descricao para materia e imprimindo
nome_estd.rename(columns={'descricao': 'materia'})
```

## Agrupamento da média de notas por matéria por turma

```python
#Juntando a tabela de notas com a tabela dos estudantes
tabela_nota_join_estudante = tabela_nota.merge(tabela_estudante, left_on = 'id_estudante', right_on = 'id') 

#Juntando a tabela anterior com a tabela das materias
tabela_nota_join_estudante = tabela_nota_join_estudante.merge(tabela_materia, left_on = 'id_materia', right_on = 'id')

#Agrupando a media das notas por materia por turma
tabela_nota_join_estudante.groupby(['descricao', 'nro_turma']).nota.agg(['mean']).round(decimals = 2)
```
